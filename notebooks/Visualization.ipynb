```python
def plot_lockdown(df, labels):
    start_date, end_date = '2020-03-25', '2020-05-31'
    mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    
    start_date_phase1, end_date_phase1 = '2020-03-25', '2020-04-14'
    m1 = (df['Date'] >= start_date_phase1) & (df['Date'] <= end_date_phase1)

    start_date_phase2, end_date_phase2 = '2020-04-14', '2020-05-03'
    m2 = (df['Date'] >= start_date_phase2) & (df['Date'] <= end_date_phase2)

    start_date_phase3, end_date_phase3 = '2020-05-03', '2020-05-17'
    m3 = (df['Date'] >= start_date_phase3) & (df['Date'] <= end_date_phase3)

    start_date_phase4, end_date_phase4 = '2020-05-17', '2020-05-31'
    m4 = (df['Date'] >= start_date_phase4) & (df['Date'] <= end_date_phase4)
    
    fig = go.Figure()
    
    fig.add_trace(go.Scatter(x = df.loc[m1].Date, y = df.loc[m1][labels], mode = 'lines', name = 'Lockdown Phase 1'))
    fig.add_trace(go.Scatter(x = df.loc[m2].Date, y = df.loc[m2][labels], mode = 'lines', name = 'Lockdown Phase 2'))
    fig.add_trace(go.Scatter(x = df.loc[m3].Date, y = df.loc[m3][labels], mode = 'lines', name = 'Lockdown Phase 3'))
    fig.add_trace(go.Scatter(x = df.loc[m4].Date, y = df.loc[m4][labels], mode = 'lines', name = 'Lockdown Phase 4'))

    fig.add_vline(x = '2020-04-14', line_dash = 'dash') # Start of phase 2

    fig.add_vline(x = '2020-05-03', line_dash = 'dash') # Start of phase 3

    fig.add_vline(x = '2020-05-17', line_dash = 'dash') # Start of phase 4
    
    fig.add_annotation(
        x = '2020-04-05',
        y = '8500',
        text = 'Lockdown Phase 1',
        font = dict(
        family = "Courier New, monospace", size = 14, color = 'Black'),
        showarrow = False
    )
    
    fig.add_annotation(
        x = '2020-04-24',
        y = '8500',
        text = 'Lockdown Phase 2',
        font = dict(
        family = "Courier New, monospace", size = 14, color = 'Black'),
        showarrow = False
    )
    
    fig.add_annotation(
        x = '2020-05-10',
        y = '8500',
        text = 'Lockdown Phase 3',
        font = dict(
        family = "Courier New, monospace", size = 14, color = 'Black'),
        showarrow = False
    )
    
    fig.add_annotation(
        x = '2020-05-24',
        y = '8500',
        text = 'Lockdown Phase 4',
        font = dict(
        family = "Courier New, monospace", size = 14, color = 'Black'),
        showarrow = False
    )

    fig.show()
    #fig.write_html("Lockdown_Impact.html")    
```

```python
#Plots mean new cases per day in the specified period, state-wise
# Decent color scales : deep, Oranges


def get_map_plot(data, geojson, locations, field, start_date,end_date, hover_name = None, title = None, hover_data = None, color_scale = 'deep'):
    
    mask = (data['Date'] >= start_date) & (data['Date'] <= end_date)
    df_temp = data.loc[mask]
    
    #Need to calculate mean counts during this period
    mean_vals = []
    for state in df_temp['State'].unique():
        temp = df_temp[df_temp['State'] == state]
        key_id = temp['id'].unique().item()
        temp = temp[field]
        mean_vals.append([state, temp.mean(), key_id])
        
    #print(mean_vals)    
    df_temp = pd.DataFrame(mean_vals, columns = ['State', field, 'id'])
    #print(df_temp)
    #print(df_temp)
    #print(df_temp.mean()['new cases'])
    #print(df_temp.head())
    mean_val = None
    if color_scale not in ['deep']:
        mean_val = 0
        
    fig = px.choropleth(
        df_temp,
        locations = locations,
        geojson = geojson,
        color = field,
        color_continuous_scale = color_scale,
        hover_name = hover_name,
        hover_data = hover_data,
        title = title
    )
    

    
    fig.update_layout(
    title={
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})
    
    fig.update_layout(coloraxis_colorbar=dict(
    yanchor="top", y=1,
    xanchor = "right", x = 0.8,
))
    
    fig.update_geos(fitbounds = "locations", visible = False)
    return fig
```

```python
#Barddhaman missing, Paschim Medinipur

wb_districts['District'].unique()
```

```python
wb_districts['id'] = wb_districts['District'].apply(lambda x: wb_id_map[x])
```

```python
start_date, end_date = '2021-03-01', '2021-04-14'

mask = (wb_districts['Date'] >= start_date) & (wb_districts['Date'] <= end_date)
df_temp = wb_districts.loc[mask]

#Need to calculate mean counts during this period
mean_vals = []
for district in df_temp['District'].unique():
    temp = df_temp[df_temp['District'] == district]
    key_id = temp['id'].unique().item()
    temp = temp['Confirmed']
    mean_vals.append([district, temp.mean(), key_id])

#print(mean_vals)    
df_temp = pd.DataFrame(mean_vals, columns = ['District', 'Confirmed', 'id'])
#print(df_temp)
#print(df_temp)
#print(df_temp.mean()['new cases'])
#print(df_temp.head())
#mean_val = None
#if color_scale not in ['deep']:
#    mean_val = 0

fig = px.choropleth(
    df_temp,
    locations = 'id',
    geojson = wb_geojson,
    color = 'Confirmed',
    hover_name = 'District',
    hover_data = ['District', 'Confirmed', 'id'],
    title = 'West Bengal District-Wise Mean Counts'
)



fig.update_layout(
title={
    'y':0.9,
    'x':0.5,
    'xanchor': 'center',
    'yanchor': 'top'})

fig.update_layout(coloraxis_colorbar=dict(
yanchor="top", y=1,
xanchor = "right", x = 1,
))

fig.update_geos(fitbounds = "locations", visible = False)
```
